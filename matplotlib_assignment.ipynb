#5. Axis, Limits, and Scaling

```python
# Define
x = np.linspace(-10, 10)
y = 1 / (x**2 + 1)

#Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = 1 / (x^2 + 1)")
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.grid(True)
plt.show()

# reversed axes
plt.figure(figsize=(8, 6))
plt.plot(y, x)
plt.xlabel("f(x)")
plt.ylabel("x")
plt.title("Reversed Axes")
plt.xlim(0, 1)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()

# plt.axis('tight')
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title("Tight Plot")
plt.axis('tight')
plt.show()


# plt.axis('equal')
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title("Equal plot")
plt.axis('equal')
plt.show()
```

#6. Labels and Legends

```python
# Define
x = np.linspace(0, 10)
y_f = np.sqrt(x)
y_g = x**(1/3)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label=('sqrt(x)')
plt.plot(x, y_g, label='x^(1/3)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = sqrt(x) and g(x) = x^(1/3)')
plt.legend()
plt.grid(True)
plt.show()
```

#7. Basic Scatter Plot

```python
#Define
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# plt.plot()
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot using plt.plot()")
plt.show()

# plt.scatter()
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot using plt.scatter()")
plt.show()
```

#8. Customizing Scatter Plots

```python
#Define
x = np.linspace(0, 4 * np.pi)
y = x * np.sin(x)

# plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=50,
            c='skyblue',
            edgecolors='darkblue',
            linewidths=1.5,
            marker='o')
plt.plot(x, y, linewidth=2, color='darkblue')
plt.xlabel("x")
plt.ylabel("f(x) = x * sin(x)")
plt.title("Scatter Plot of f(x) = x * sin(x)")
plt.show()
```

#9. Custom Plot with Alternative Style

```python
#Define
x = np.linspace(-5, 5)
y_f = np.tan(x)
y_g = x * np.exp(-x**2)
y_h =np.sin(x) / x
y_z = np.arctan(x)

#plot
plt.style.use('ggplot')
plt.figure(figsize=(12, 8))
plt.plot(x, y_f, color='skyblue', linestyle='-', linewidth=2, label=r'f(x)')
plt.plot(x, y_g, color='orange', linestyle='--', linewidth=2, label=r'g(x)')
plt.plot(x, y_h, color='green', linestyle=':', linewidth=2, label=r'h(x)')
plt.plot(x, y_z, color='purple', linestyle='-.', linewidth=2, label=r'z(x)')
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("Plot of Multiple Functions", fontsize=16)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend(fontsize=12)
plt.show()
```

#10. Custom Plot with Alternative Style

```python
#Define
x = np.linspace(-5, 5, 400)
y_f = np.tan(x)
y_g =x * np.exp(-x**2)
y_h = np.sin(x) / x
y_z = np.arctan(x)

#I Chose these functions becaus e they were esiaer to type into the program and trigeronometric function have a elegant beauty espeecially overlayed


#Plot Style 1
plt.style.use('default')
plt.figure(figsize=(12, 8))
plt.plot(x, y_f, label=r'f(x)')
plt.plot(x, y_g, label=r'g(x)')
plt.plot(x, y_h, label=r'h(x)')
plt.plot(x, y_z, label=r'z(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Plot Style default")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)
plt.show()

#Plot Style 2
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 8))
plt.plot(x, y_f, label=r'f(x)')
plt.plot(x, y_g, label=r'g(x')
plt.plot(x, y_h, label=r'h(x)')
plt.plot(x, y_z, label=r'z(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Plot Style Seaborn-Whitegrid")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)
plt.show()

#Plot Syle 3
#I personally like the classic because the gridlines are less intrusive being dotted and the graphic is not as big so it
#fits on my screen a bit nicer
plt.style.use('classic')
plt.figure(figsize=(12, 8))
plt.plot(x, y_f, label=r'f(x)')
plt.plot(x, y_g, label=r'g(x)')
plt.plot(x, y_h, label=r'h(x)')
plt.plot(x, y_z, label=r'z(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Plot Style classic")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)
plt.show()
```

#11. Style Comparison

```python
#Choosing plot created in Task 2. and using the seaborn-whitegrid and the classic style

#Seaborn-Whitegrid plot
x = np.linspace(-2, 4)
y = x**2 - 2*x + 1
plt.plot(x, y)
plt.style.use('seaborn-v0_8-whitegrid')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of function")
plt.grid(True)
plt.show()

#Classic plot
x = np.linspace(-2, 4)
y = x**2 - 2*x + 1
plt.style.use('classic')
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of function")
plt.grid(True)
plt.show()
```

This is a tough comparison, they both are pleasing to the eye but if I had to choose, I would go with classic. If I were just casually veiwing data plots, and only needed to know the gist of the data without fully understanding the depth of the plot the 'seaborn-whitegrid' would work. It's sleek, well put together and the contrast of the background of the graph is very pleasing to the eye. However those white lines make is slightly harder to see precise points and follow the grid to an exact point. The Classic gridliines and dotted, making them non-imposing, buts still present to follow to an exact point. I would want to use classic if interpreting the graph is necessary for understanding i.e. in a textbook or mathematical reference book.
